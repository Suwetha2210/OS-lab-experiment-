```python
%%bash

# Assign a number directly for demonstration. You can change this value.
n=5

# Optionally, you can uncomment the lines below if you expect interactive input
# echo "ENTER THE NUMBER:"
# read n

fact=1

while [ $n -gt 1 ]
do
    fact=$((fact * n))
    n=$((n - 1))
done

echo "FACTORIAL OF THE GIVEN NUMBER IS $fact"
```
