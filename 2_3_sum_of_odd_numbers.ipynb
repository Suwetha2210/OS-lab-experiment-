```python
%%bash

#!/bin/bash

# Program to find the sum of odd numbers within the given range

echo "ENTER THE RANGE:"
read n

x=1
sum=0

while [ $x -le $n ]
do
    sum=$((sum + x))
    x=$((x + 2))
done

# Display the result
echo "SUM = $sum"
```
